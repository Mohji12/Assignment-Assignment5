```python
from sklearn.linear_model import LinearRegression

Lr = LinearRegression()
Lr.fit(x_train,y_train)
```

```python
# Ridge regressor Regularization technique

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge

cross_val_Rigde = []

alpha = []

for i in range(1,15):
    model = Ridge(alpha=i*0.25)
    result = cross_val_score(model,x_train,y_train,cv=10)
    avg_cross_val_score = mean(result)*100
    cross_val_Rigde.append(avg_cross_val_score)
    alpha.append(i * 0.25)

for i in  range(0,len(alpha)):
    print(str(alpha[i])+ ' : '+str(cross_val_Rigde[i]))

# From our above code the best value for alpha is 3.5
```

```python
# testing the model on testing dataset
from sklearn.metrics import r2_score

model_ridge = Ridge(alpha=3)
model_ridge.fit(x_train,y_train)

y_pred_ridge = model_ridge.predict(x_test)
r2_score(y_test,y_pred_ridge)
```
