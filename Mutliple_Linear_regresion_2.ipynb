```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.3)
```

```python
x_train
```

```python
x_train = pd.get_dummies(x_train,dtype="int32")
x_train
```

```python
x_test = pd.get_dummies(x_test,dtype="int32")
x_test
```

```python
col = x_train.columns
```

```python

from sklearn.preprocessing import StandardScaler

x_train = pd.DataFrame(StandardScaler().fit_transform(x_train),columns=col)
x_train.head()
```

```python
x_test = pd.DataFrame(StandardScaler().fit_transform(x_test),columns=col)
x_test
```

```python
from sklearn.linear_model import LinearRegression

Lr = LinearRegression()
Lr.fit(x_train,y_train)
```

```python
Lr.intercept_
```

```python
Lr.coef_
```

```python
# training Accuracy

from sklearn.metrics import r2_score,mean_absolute_error

print(r2_score(y_train,Lr.predict(x_train)))
mean_absolute_error(y_train,Lr.predict(x_train))
```

```python
# testing Accuracy

print(r2_score(y_test,Lr.predict(x_test)))
print(mean_absolute_error(y_test,Lr.predict(x_test)))
```

```python
import seaborn as sns

sns.regplot(x = y_test,y = Lr.predict(x_test))

```

```python
# Ridge regressor Regularization technique

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge

cross_val_Rigde = []

alpha = []

for i in range(1,15):
    model = Ridge(alpha=i*0.25)
    result = cross_val_score(model,x_train,y_train,cv=10)
    avg_cross_val_score = mean(result)*100
    cross_val_Rigde.append(avg_cross_val_score)
    alpha.append(i * 0.25)

for i in  range(0,len(alpha)):
    print(str(alpha[i])+ ' : '+str(cross_val_Rigde[i]))

# From our above code the best value for alpha is 3
```

```python
# testing the model on testing dataset
from sklearn.metrics import r2_score

model_ridge = Ridge(alpha=3)
model_ridge.fit(x_train,y_train)

y_pred = model_ridge.predict(x_test)
r2_score(y_test,y_pred)
```

```python
# Lasso regressor Regularization technique

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Lasso

cross_val_Rigde = []

alpha = []

for i in range(1,15):
    model = Lasso(alpha=i*0.25)
    result = cross_val_score(model,x_train,y_train,cv=10)
    avg_cross_val_score = mean(result)*100
    cross_val_Rigde.append(avg_cross_val_score)
    alpha.append(i * 0.25)

for i in  range(0,len(alpha)):
    print(str(alpha[i])+ ' : '+str(cross_val_Rigde[i]))

# From our above code the best value for alpha is 2.5
```

```python
# testing the model_Lasso on testing dataset
from sklearn.metrics import r2_score

model_ridge = Lasso(alpha=2.5)
model_ridge.fit(x_train,y_train)

y_pred = model_ridge.predict(x_test)
r2_score(y_test,y_pred)
```

```python
# Accuracy for Tesitng
data1 = {'LinearRegression_Accuracy':r2_score(y_train,Lr.predict(x_train)),
                           'ridge_regression_Accuracy':r2_score(y_test,y_pred),
                           'Lasso_regressor_Accuracy':r2_score(y_test,y_pred)}
print(data1.keys(),"\nb",data1.values())
```
